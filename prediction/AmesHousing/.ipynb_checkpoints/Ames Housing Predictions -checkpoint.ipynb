## Read in the data

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
renaming = {
    'MSSubClass' : 'BuildingClass',
    'MSZoning' : 'ZoningClass',
    'LotFrontage' : 'AmountOfStreet',
    'Street' : 'RoadAccessType',
    'Alley' : 'AlleyAccessType',
    'LandContour' : 'Flatness',
    'Condition1' : 'ProximityToRoadRail1',
    'Condition2' : 'ProximityToRoadRail2',
    'OverallQual' : 'MaterialQuality',
    'OverallCond' : 'ConditionRating',
    'YearRemodAdd' : 'RemodelDate',
    'RoofStyle' : 'RoofType',
    'RoofMatl' : 'RoofMaterial',
    'Heating' : 'HeatingType',
    'HeatingQC' : 'HeatingQualityCond',
    'LowQualFinSF' : 'LowQualityFinshedSF',
    'Functional' : 'HomeFunctionalRating',
    'FireplaceQu' : 'FireplaceQuality',
    '3SsnPorch' : 'ThreeSeasonPorchArea',
    'PoolQC' : 'PoolQualityCond',
    'MoSold' : 'MonthSold',
    'YrSold' : 'YearSold',
    'Fence' : 'FenceQuality'
}

train.rename(index=str, inplace=True, columns=renaming)
test.rename(index=str, inplace=True, columns=renaming)
```

## Get baseline model as quick as possible

```python
train.head()
```

```python
train.shape
```

```python
categorical_features = train.select_dtypes(include = ["object"]).columns
numerical_features = train.select_dtypes(exclude = ["object"]).columns
# Drop all columns with N/A values
train_no_na = train.dropna(axis=1)
# Remove the target variable 
y = train_no_na['SalePrice']
train_no_na = train_no_na.drop(labels=['SalePrice'], axis=1)
# Encode categorical variables 
from sklearn import preprocessing
for column in categorical_features:
    if column in train_no_na:
        le = preprocessing.LabelEncoder()
        train_no_na[column] = le.fit_transform(train_no_na[column])
```

## Now that basic preprocessing is done, we are ready to build a baseline model

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_baseline = RandomForestRegressor(n_estimators=1000, oob_score=True, n_jobs=-1, random_state=42)
random_forest_baseline.fit(train_no_na, y)
```

```python
# From the docs, The RandomForestClassifier is trained using bootstrap aggregation, where each new tree is fit from a 
# bootstrap sample of the training observations z_i = (x_i, y_i). The out-of-bag (OOB) error is the average error for 
# each z_i calculated using predictions from the trees that do not contain z_i in their respective bootstrap sample
# The study of error estimates for bagged classifiers in Breiman [1996b], gives empirical evidence to show that the out-of-bag 
# estimate is as accurate as using a test set of the same size as the training set. Source: https://www.quora.com/What-is-the-out-of-bag-error-in-random-forests-What-does-it-mean-Whats-a-typical-value-if-any-Why-would-it-be-higher-or-lower-than-a-typical-value
random_forest_baseline.oob_score_
from sklearn import metrics
y_oob = random_forest_baseline.oob_prediction_
# For this metric, the higher the number (max is 1), the better the predictions
print('explained_variance_score = ', metrics.explained_variance_score(y, y_oob))
# A total of 17597 dollars off over the entire training set, not bad
print('mean_absolute_error = ', metrics.mean_absolute_error(y, y_oob))
```

```python
# We will use these feature importances as a baseline later on when we do some feature engineering
baseline_feature_importances = eda.feature_importances(random_forest_baseline, train_no_na)
```

```python
missing_df = eda.missing_values(train)
missing_df = eda.missing_values(test)
```

## Now that we have a baseline model, let's set about improving it. We will attempt to do this via intelligently imputing the data, training only on important features, feature engineering (to some extent), finding optimal hyperparameters for the model, and possibly trying out other models as well. First, let's impute the data

```python
for dataset in [train, test]:
    # N/A values for this features mean the the feature doesn't exist for that house
    # Later we will use LabelEncoder to encode these strings 
    dataset['AlleyAccessType'].fillna("No Alley", inplace=True)
    dataset['MiscFeature'].fillna("No misc feature", inplace=True)
    dataset['PoolQualityCond'].fillna("No pool", inplace=True)
    dataset['FenceQuality'].fillna('No fence', inplace=True)
    dataset['FireplaceQuality'].fillna('No fireplace', inplace=True)
    # Assume N/A values of AmountOfStreet means no access street in front of house
    dataset['AmountOfStreet'].fillna(0.0, inplace=True)
    # Assume that N/A values for these features indicate there is no garage for the house
    # Each of these features have same # of NAN values
    garage_ft = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']
    garage_numerical_features = ['GarageCars', 'GarageArea']
    for feature in garage_ft:
        dataset[feature].fillna('No garage', inplace=True)
        
    for feature in garage_numerical_features:
        dataset[feature].fillna(0, inplace=True)
    
    # Fill NA year with negative value
    dataset['GarageYrBlt'].fillna(-1, inplace=True)

    basement_ft = ['BsmtQual','BsmtCond','BsmtExposure','BsmtFinType2', 'BsmtFinType1']
    basement_sq_ft = ['BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'BsmtHalfBath', 'BsmtFullBath']
    # NAN values for this feature indicate no basement
    # Other basement square footage features are zero where the above basement features are nan
    for feature in basement_ft:
        dataset[feature].fillna('No basement', inplace=True)
        
    for feature in basement_sq_ft:
        dataset[feature].fillna(0, inplace=True)
    
    # NAN values for this feature indicate no nmasonry veneer
    dataset['MasVnrType'].fillna('No masonry veneer', inplace=True)
    dataset['MasVnrArea'].fillna(0, inplace=True)
    
    # The lone house with no electricity
    dataset['Electrical'].fillna('No electrical', inplace=True)

test['ZoningClass'].fillna('No Zoning Class', inplace=True)
test['Utilities'].fillna('No utilities', inplace=True)
test['Exterior1st'].fillna('No Exterior1st', inplace=True)
test['Exterior2nd'].fillna('No Exterior2nd', inplace=True)
test['SaleType'].fillna('Oth', inplace=True)
test.isnull().any().sort_values(ascending=False)

le = preprocessing.LabelEncoder()
kitchen_qual_median = pd.Series(le.fit_transform(test['KitchenQual'].dropna())).median() # equals 3, which maps to 'TA'
test['KitchenQual'].fillna('TA', inplace=True)

pd.Series(le.fit_transform(test['HomeFunctionalRating'].dropna())).median() # 6, which maps to 'Typ'
test['HomeFunctionalRating'].fillna('Typ', inplace=True)
print(test.isnull().any().sort_values(ascending=False).any())
print(train.isnull().any().sort_values(ascending=False).any())
# We've cleared up all NANs!!!
```
